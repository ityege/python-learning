### 3.6.3 用函数 step()绘制阶梯图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, 10)
y = np.sin(x)
# def step(
#     x: ArrayLike,
#     y: ArrayLike,
#     *args,
#     where: Literal["pre", "post", "mid"] = "pre",
#     data=None,
#     **kwargs,
# ) -> list[Line2D]:

# x：表示阶梯图的 x 坐标值，即横坐标。这可以是一个包含实数值的数组或序列。
# y：表示阶梯图的 y 坐标值，即纵坐标。这可以是一个包含实数值的数组或序列。
# color="#8dd3c7"：设置阶梯图的颜色。这里使用的颜色是 "#8dd3c7"，表示一种自定义的颜色。
# where="pre"：设置阶梯图线段的位置。这里设置为 "pre"，表示线段的起点位于相应的 x 坐标处。
# lw=2：设置线宽为 2。这里将阶梯图的线宽设置为 2。
plt.step(x, y, color="#8dd3c7", where="pre", lw=2)
plt.xlim(0, 11)
plt.xticks(np.arange(1, 11, 1))
plt.ylim(-1.2, 1.2)
plt.show()
```

### 3.7.5 直方图的不同形状

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
import matplotlib.pyplot as plt
import numpy as np

# set test scores
scoresT1 = np.random.randint(0, 100, 100)
scoresT2 = np.random.randint(0, 100, 100)
x = [scoresT1, scoresT2]
colors = ["#8dd3c7", "#bebada"]
labels = ["班级 A", "班级 B"]
# plot histogram
bins = range(0, 101, 10)
# def hist(
#     x: ArrayLike | Sequence[ArrayLike],
#     bins: int | Sequence[float] | str | None = None,
#     range: tuple[float, float] | None = None,
#     density: bool = False,
#     weights: ArrayLike | None = None,
#     cumulative: bool | float = False,
#     bottom: ArrayLike | float | None = None,
#     histtype: Literal["bar", "barstacked", "step", "stepfilled"] = "bar",
#     align: Literal["left", "mid", "right"] = "mid",
#     orientation: Literal["vertical", "horizontal"] = "vertical",
#     rwidth: float | None = None,
#     log: bool = False,
#     color: ColorType | Sequence[ColorType] | None = None,
#     label: str | Sequence[str] | None = None,
#     stacked: bool = False,
#     *,
#     data=None,
#     **kwargs,
# ) -> tuple[
#     np.ndarray | list[np.ndarray],
#     np.ndarray,
#     BarContainer | Polygon | list[BarContainer | Polygon],
# ]:

# x：表示要绘制直方图的数据。这可以是一个包含实数值的数组或序列。
# bins：表示直方图的箱体数量或边界。这可以是一个整数，表示箱体数量，或一个包含箱体边界值的数组。
# color：设置直方图的颜色。这里使用的颜色可以是一个字符串，表示所有直方图的颜色，或者是一个包含颜色值的数组，表示每个直方图的颜色。
# histtype：设置直方图的类型。这里设置为 "bar"，表示绘制垂直的条形直方图。
# rwidth：设置直方图条形的宽度。这里将条形的宽度设置为 10。
# stacked：设置是否堆叠直方图。这里设置为 True，表示堆叠直方图。
# label：设置直方图的标签。这可以是一个字符串，表示所有直方图的标签，或者是一个包含标签的数组，表示每个直方图的标签。
plt.hist(x,
         bins=bins,
         color=colors,
         histtype="bar",
         rwidth=10,
         stacked=True,
         label=labels)
# set x,y-axis label
plt.xlabel("测试成绩")
plt.ylabel("学生人数")
plt.title("不同班级的测试成绩的直方图")
plt.legend(loc="upper left")
plt.show()
```

## 3.8 饼图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
labels = "A 难度水平", "B 难度水平", "C 难度水平", "D 难度水平"
students = [0.35, 0.15, 0.20, 0.30]
colors = ["#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
explode = (0.1, 0.1, 0.1, 0.1)
# exploded pie chart
# def pie(
#     x: ArrayLike,
#     explode: ArrayLike | None = None,
#     labels: Sequence[str] | None = None,
#     colors: ColorType | Sequence[ColorType] | None = None,
#     autopct: str | Callable[[float], str] | None = None,
#     pctdistance: float = 0.6,
#     shadow: bool = False,
#     labeldistance: float | None = 1.1,
#     startangle: float = 0,
#     radius: float = 1,
#     counterclock: bool = True,
#     wedgeprops: dict[str, Any] | None = None,
#     textprops: dict[str, Any] | None = None,
#     center: tuple[float, float] = (0, 0),
#     frame: bool = False,
#     rotatelabels: bool = False,
#     *,
#     normalize: bool = True,
#     hatch: str | Sequence[str] | None = None,
#     data=None,
# ) -> tuple[list[Wedge], list[Text]] | tuple[list[Wedge], list[Text], list[Text]]:


# students：表示用于绘制饼图的数据。这可以是一个包含数值的数组或序列。
# explode：设置饼图各个扇区的偏移量。这可以是一个数组，表示每个扇区的偏移量。
# labels：设置饼图各个扇区的标签。这可以是一个包含字符串的数组，表示每个扇区的标签。
# autopct="%3.1f%%"：设置饼图中扇区的百分比显示格式。这里使用 %3.1f%% 表示显示一位小数的百分比。
# startangle=45：设置饼图的起始角度。这里将起始角度设置为 45 度。
# shadow=True：设置是否显示饼图的阴影效果。
# colors：设置饼图各个扇区的颜色。这可以是一个包含颜色值的数组，表示每个扇区的颜色。
plt.pie(students,
        explode=explode,
        labels=labels,
        autopct="%3.1f%%",
        startangle=45,
        shadow=True,
        colors=colors)
plt.title("选择不同难度测试试卷的学生百分比")
plt.show()
```

### 3.8.4 绘制内嵌环形饼图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
elements = ["面粉", "砂糖", "奶油", "草莓酱", "坚果"]
weight1 = [40, 15, 20, 10, 15]
weight2 = [30, 25, 15, 20, 10]
colormapList = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
outer_colors = colormapList
inner_colors = colormapList
# weight1 和 weight2：表示用于绘制饼图的数据。这可以是一个包含数值的数组或序列。
# autopct="%3.1f%%"：设置饼图中扇区的百分比显示格式。这里使用 %3.1f%% 表示显示一位小数的百分比。
# radius：设置饼图的半径大小。radius=1 表示第一个饼图的半径是整个饼图的半径，radius=0.7 表示第二个饼图的半径是整个饼图半径的 0.7 倍。
# pctdistance：设置百分比标签的距离中心的位置。pctdistance=0.85 表示第一个饼图的百分比标签距离中心的位置是半径的 0.85 倍，pctdistance=0.75 表示第二个饼图的百分比标签距离中心的位置是半径的 0.75 倍。
# colors：设置饼图各个扇区的颜色。这可以是一个包含颜色值的数组，表示每个扇区的颜色。
# textprops=dict(color="w")：设置文本标签的属性。这里将文本颜色设置为白色。
# wedgeprops=dict(width=0.3, edgecolor="w")：设置扇区的属性。这里将扇区的宽度设置为 0.3，边缘颜色设置为白色。
wedges1, texts1, autotexts1 = plt.pie(weight1,
                                      autopct="%3.1f%%",
                                      radius=1,
                                      pctdistance=0.85,
                                      colors=outer_colors,
                                      textprops=dict(color="w"),
                                      wedgeprops=dict(width=0.3, edgecolor="w"))
wedges2, texts2, autotexts2 = plt.pie(weight2,
                                      autopct="%3.1f%%",
                                      radius=0.7,
                                      pctdistance=0.75,
                                      colors=inner_colors,
                                      textprops=dict(color="w"),
                                      wedgeprops=dict(width=0.3, edgecolor="w"))
# wedges1：表示要在图例中显示的饼图对象。这里将第一个饼图的扇区对象传递给图例。
# elements：表示每个图例项的名称。这可以是一个包含图例项名称的列表或数组。
# fontsize=12：设置图例文本的字体大小。
# title="配料表"：设置图例的标题文本。
# loc="center left"：设置图例的位置。这里将图例放置在图形的左侧中心。
# bbox_to_anchor=(0.91, 0, 0.3, 1)：设置图例的边界框位置和大小。这里的参数表示图例边界框的左下角位置为 (0.91, 0)，宽度为 0.3，高度为 1。
plt.legend(wedges1,
           elements,
           fontsize=12,
           title="配料表",
           loc="center left",
           bbox_to_anchor=(0.91, 0, 0.3, 1))
plt.setp(autotexts1, size=15, weight="bold")
plt.setp(autotexts2, size=15, weight="bold")
plt.setp(texts1, size=12)
plt.title("不同果酱面包配料比例表的比较")
plt.show()
```

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"] = ["LiSu"]
mpl.rcParams["axes.unicode_minus"] = False
# some simple data
x = np.arange(5)
y = [1200, 2400, 1800, 2200, 1600]
std_err = [150, 100, 180, 130, 80]
bar_width = 0.6
colors = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
# create horizontal bar
plt.barh(x, y,
         bar_width,
         color=colors,
         align="center",
         xerr=std_err,
         tick_label=["家庭", "小说", "心理", "科技", "儿童"])
# set x,y_axis label
plt.xlabel("订购数量")
plt.ylabel("图书种类")
# set title
plt.title("大型图书展销会的不同图书种类的采购情况")
# set xaxis grid
plt.grid(True, axis="x", ls=":", color="gray", alpha=0.2)
plt.xlim(0, 2600)
plt.show()
```

### 带误差棒的堆积柱状图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"] = ["LiSu"]
mpl.rcParams["axes.unicode_minus"] = False
# some simple data
x = np.arange(5)
y1 = [1200, 2400, 1800, 2200, 1600]
y2 = [1050, 2100, 1300, 1600, 1340]
std_err1 = [150, 100, 180, 130, 80]
std_err2 = [120, 110, 170, 150, 120]
bar_width = 0.6
tick_label = ["家庭", "小说", "心理", "科技", "儿童"]
error_attri = dict(ecolor="black", elinewidth=2, capsize=0)
# create bar
plt.bar(x, y1,
        bar_width,
        color="#6495ED",
        align="center",
        yerr=std_err1,
        label="地区 1",
        error_kw=error_attri)
plt.bar(x, y2,
        bar_width,
        bottom=y1,
        color="#FFA500",
        align="center",
        yerr=std_err2,
        label="地区 2",
        error_kw=error_attri)
# set x,y_axis label
plt.xlabel("图书种类")
plt.ylabel("订购数量")
# set title
plt.title("不同地区大型图书展销会的图书采购情况")
# set yaxis grid
plt.grid(True, axis="y", ls=":", color="gray", alpha=0.2)
plt.xticks(x, tick_label)
plt.legend()
plt.show()
```
