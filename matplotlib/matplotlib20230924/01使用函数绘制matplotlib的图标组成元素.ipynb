### 1.3 绘制 matplotlib 图表组成元素的函数用法

#### 1.3.1 函数 plot()——展现变量的趋势变化

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05, 10, 1000)
y = np.cos(x)
# @_copy_docstring_and_deprecators(Axes.plot)
# def plot(
#     *args: float | ArrayLike | str,
#     scalex: bool = True,
#     scaley: bool = True,
#     data=None,
#     **kwargs,
# ) -> list[Line2D]:
# '- _' is not a valid value for ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x, y, ls="-", lw=1, label="plot figure")
plt.legend()  # 显示图例,右上角对象
plt.show()#显示图像
```

#### 1.3.2 函数 scatter()——寻找变量之间的关系

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.random.rand(1000)
# def scatter(
#     x: float | ArrayLike,
#     y: float | ArrayLike,
#     s: float | ArrayLike | None = None,
#     c: Sequence[ColorType] | ColorType | None = None,
#     marker: MarkerType | None = None,
#     cmap: str | Colormap | None = None,
#     norm: str | Normalize | None = None,
#     vmin: float | None = None,
#     vmax: float | None = None,
#     alpha: float | None = None,
#     linewidths: float | Sequence[float] | None = None,
#     *,
#     edgecolors: Literal["face", "none"] | ColorType | Sequence[ColorType] | None = None,
#     plotnonfinite: bool = False,
#     data=None,
#     **kwargs,
# ) -> PathCollection:
plt.scatter(x,y,label="scatter figure")
plt.legend()
plt.show()
```

#### 1.3.3 函数 xlim()——设置 x 轴的数值显示范围

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05,10,1000)
y = np.random.rand(1000)
plt.scatter(x,y,label="scatter figure")
plt.legend()
plt.xlim(0.05,10)
plt.ylim(0,1)
plt.show()
```

#### 1.3.4 函数 xlabel()——设置 x 轴的标签文本

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05,10,1000)
y = np.sin(x)
plt.plot(x,y,ls="-.",lw=2,c="c",label="plot figure")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

#### 1.3.5 函数 grid()——绘制刻度线的网格线

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.sin(x)
plt.plot(x,y,ls="-.",lw=2,c="c",label="plot figure")
plt.legend()
# def grid(
#     visible: bool | None = None,
#     which: Literal["major", "minor", "both"] = "major",
#     axis: Literal["both", "x", "y"] = "both",
#     **kwargs,
# ) -> None:
plt.grid(linestyle=":",color="r")
plt.show()
```

#### 1.3.6 函数 axhline()——绘制平行于 x 轴的水平参考线

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.sin(x)
plt.plot(x,y,ls="-.",lw=2,c="c",label="plot figure")
plt.legend()
# def axhline(y: float = 0, xmin: float = 0, xmax: float = 1, **kwargs) -> Line2D:
plt.axhline(y=0.0,c="r",ls="--",lw=2)#绘制平行于 x 轴的水平参考线。
plt.axvline(x=4.0,c="r",ls="--",lw=2)
plt.show()
```

#### 1.3.7 函数 axvspan()——绘制垂直于 x 轴的参考区域

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.sin(x)
plt.plot(x,y,ls="-.",lw=2,c="c",label="plot figure")
plt.legend()
# def axvspan(
#     xmin: float, xmax: float, ymin: float = 0, ymax: float = 1, **kwargs
# ) -> Polygon:
plt.axvspan(xmin=4.0,xmax=6.0,facecolor="y",alpha=0.3)
plt.axhspan(ymin=0.0,ymax=0.5,facecolor="y",alpha=0.3)
plt.show()
```

#### 1.3.8 函数 annotate()——添加图形内容细节的指向型注释文本

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.05,10,1000)
y = np.sin(x)
plt.plot(x,y,ls="-.",lw=2,c="c",label="plot figure")
plt.legend()
plt.annotate(
    "maximum",                         # 注释文本内容为 "maximum"
    xy=(np.pi/2, 1.0),                  # 箭头指向的点的坐标为 (np.pi/2, 1.0)
    xytext=((np.pi/2) + 1.0, 0.8),      # 注释文本的起始坐标为 ((np.pi/2) + 1.0, 0.8)
    weight="bold",                      # 注释文本的字体粗细为 "bold"
    color="b",                          # 注释文本的颜色为蓝色 ("b")
    arrowprops=dict(
        arrowstyle="->",                 # 箭头的样式为 "->"
        connectionstyle="arc3",          # 箭头连接线的样式为弧形
        color="b"                        # 箭头的颜色为蓝色 ("b")
    )
)
plt.show()
```

#### 1.3.9 函数 text()——添加图形内容细节的无指向型注释文本

```python
import matplotlib.pyplot as plt
import numpy as np

# define data
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
y1 = np.random.randn(100)
# scatter figure
plt.scatter(x, y1, c="0.25", label="scatter figure")
# plot figure
plt.plot(x, y, ls="--", lw=2, label="plot figure")
# some clean up(removing chartjunk)
# turn the top spine and the right spine off
# 去除 Matplotlib 图形的顶部和右侧的轴脊柱（spines）
for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")

# turn bottom tick for x-axis on
plt.gca().xaxis.set_ticks_position("bottom")#用于设置 Matplotlib 图形的 x 轴刻度线的位置为底部
# set tick_line position of bottom

# leave left ticks for y-axis on
plt.gca().yaxis.set_ticks_position("left")#用于设置 Matplotlib 图形的 y 轴刻度线的位置为左
# set tick_line position of left
# set x,yaxis limit
plt.xlim(0.0, 4.0)
plt.ylim(-3.0, 3.0)
# set axes labels
plt.ylabel("y_axis")
plt.xlabel("x_axis")
# # set x,yaxis grid
plt.grid(True, ls=":", color="r")#显示网格线
# # add a horizontal line across the axis
plt.axhline(y=0.0, c="r", ls="--", lw=2)#添加横向参考线
# # add a vertical span across the axis
plt.axvspan(xmin=1.0, xmax=2.0, facecolor="y", alpha=.3)#添加纵向参考区域
# # set annotating info
# 使用 Matplotlib 的 annotate 函数在图形上添加了多个带箭头的文本注释。
# 文本内容为 "maximum"。
# 箭头指向的点的坐标为 (np.pi / 2, 1.0)。
# 注释文本的起始坐标为 ((np.pi / 2) + 0.15, 1.5)。
# 注释文本的字体粗细为 "bold"，颜色为红色。
# 箭头样式为 "->"，连接线样式为弧形，颜色为红色。
plt.annotate("maximum", xy=(np.pi / 2, 1.0),
             xytext=((np.pi / 2) + 0.15, 1.5), weight="bold", color="r",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r"))
# 文本内容为 "spines"。
# 箭头指向的点的坐标为 (0.75, -3)。
# 注释文本的起始坐标为 (0.35, -2.25)。
# 注释文本的字体粗细为 "bold"，颜色为蓝色。
# 箭头样式为 "->"，连接线样式为弧形，颜色为蓝色。
plt.annotate("spines", xy=(0.75, -3),
             xytext=(0.35, -2.25), weight="bold", color="b",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))
# 空的文本内容。
# 箭头指向的点的坐标为 (0, -2.78)。
# 注释文本的起始坐标为 (0.4, -2.32)。
# 箭头样式为 "->"，连接线样式为弧形，颜色为蓝色。
plt.annotate("", xy=(0, -2.78),
             xytext=(0.4, -2.32),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))
# 空的文本内容。
# 箭头指向的点的坐标为 (3.5, -2.98)。
# 注释文本的起始坐标为 (3.6, -2.70)。
# 箭头样式为 "->"，连接线样式为弧形，颜色为蓝色。
plt.annotate("", xy=(3.5, -2.98),
             xytext=(3.6, -2.70),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))
# set text info
# 添加无指向型注释文本
plt.text(3.6, -2.70, "'|' is tickline", weight="bold", color="b")
plt.text(3.6, -2.95, "3.5 is ticklabel", weight="bold", color="b")
# set title
plt.title("structure of matplotlib")
# set legend
plt.legend()
plt.show()
```
