#### 2.1 函数 bar()——用于绘制柱状图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt

# mpl.rcParams["font.sans-serif"] = ["SimHei"]：将 Matplotlib 的全局字体设置为 "SimHei"。这里使用的是中文字体 "SimHei"，它是一种支持中文字符的字体。通过设置全局字体，可以确保绘制的图形中的文字能够正确显示中文字符。
# mpl.rcParams["axes.unicode_minus"] = False：设置 Matplotlib 图形中显示减号（Unicode U+2212）的方式。将其设置为 False，表示不使用 Unicode 减号，而是使用 ASCII 减号（"-"）进行显示。这样做是为了解决在某些操作系统和配置中，使用 Unicode 减号可能导致减号显示异常或乱码的问题。
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
# some simple data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]
# create bar
# 绘制柱状图
# def bar(
#     x: float | ArrayLike,
#     height: float | ArrayLike,
#     width: float | ArrayLike = 0.8,
#     bottom: float | ArrayLike | None = None,
#     *,
#     align: Literal["center", "edge"] = "center",
#     data=None,
#     **kwargs,
# ) -> BarContainer:
# 参数
# x：表示柱状图的 x 坐标位置，即每个柱子的位置。这可以是一个包含实数值的数组或序列。
# y：表示柱状图的高度，即每个柱子的高度。这可以是一个包含实数值的数组或序列。
# align="center"：设置柱子的对齐方式。在这里，将柱子设置为居中对齐，即柱子的中心与 x 坐标对应的位置对齐。
# color="c"：设置柱子的颜色。这里使用的颜色是 "c"，代表青色（cyan）。
# tick_label=["q", "a", "c", "e", "r", "j", "b", "p"]：设置每个柱子的刻度标签。这里使用的标签是一个包含字符串的列表，每个字符串对应一个柱子的标签。
# hatch="/"：设置柱子的填充模式。这里使用的填充模式是斜线（"/"），使柱子的内部显示为斜线图案
tick_label = ["q", "a", "c", "e", "r", "j", "b", "p"]
plt.bar(x, y, align="center", color="c", tick_label=tick_label, hatch="/")
# set x,y_axis label
plt.xlabel("箱子编号")
plt.ylabel("箱子重量（kg）")
plt.show()
```

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
# some simple data
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
# create bar
# 绘制条形图
plt.barh(x,y,align="center",color="c",tick_label=["q","a","c","e","r","j","b","p"],hatch="/")
# set x,y_axis label
plt.xlabel("箱子重量（kg）")
plt.ylabel("箱子编号")
plt.show()
```

#### 2.3 函数 hist()——用于绘制直方图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
import matplotlib.pyplot as plt
import numpy as np

# set test scores
boxWeight: np.ndarray = np.random.randint(0, 10, 100)
x = boxWeight
# plot histogram
bins: range = range(0, 11, 1)  # [0,10]
# def hist(
#     x: ArrayLike | Sequence[ArrayLike],
#     bins: int | Sequence[float] | str | None = None,
#     range: tuple[float, float] | None = None,
#     density: bool = False,
#     weights: ArrayLike | None = None,
#     cumulative: bool | float = False,
#     bottom: ArrayLike | float | None = None,
#     histtype: Literal["bar", "barstacked", "step", "stepfilled"] = "bar",
#     align: Literal["left", "mid", "right"] = "mid",
#     orientation: Literal["vertical", "horizontal"] = "vertical",
#     rwidth: float | None = None,
#     log: bool = False,
#     color: ColorType | Sequence[ColorType] | None = None,
#     label: str | Sequence[str] | None = None,
#     stacked: bool = False,
#     *,
#     data=None,
#     **kwargs,
# ) -> tuple[
#     np.ndarray | list[np.ndarray],
#     np.ndarray,
#     BarContainer | Polygon | list[BarContainer | Polygon],
# ]:

# x：表示要绘制直方图的数据。这可以是一个包含实数值的数组或序列。
# bins：表示直方图的箱子数量或定义的箱子边界。这可以是一个整数，表示要使用的等宽箱子数量，也可以是一个表示箱子边界的数组，用于指定不等宽的箱子。
# color="g"：设置直方图的颜色。这里使用的颜色是 "g"，代表绿色（green）。
# histtype="bar"：设置直方图的类型。这里使用的类型是 "bar"，表示绘制标准的柱状图。
# rwidth=1：设置柱状图的宽度相对于箱子宽度的比例。这里将柱状图的宽度设置为与箱子宽度相等，即完全填充箱子的宽度。
# alpha=0.6：设置直方图的透明度。这里将透明度设置为 0.6，使直方图具有一定的透明效果。
plt.hist(x, bins=bins, color="g", histtype="bar", rwidth=1, alpha=0.6)
# set x,y-axis label
plt.xlabel("箱子重量（kg）")
plt.ylabel("销售数量（个）")
plt.show()
```

#### 2.4 函数 pie()——用于绘制饼图

```python
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
kinds = "简易箱", "保温箱", "行李箱", "密封箱"
colors = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3"]
soldNums = [0.05, 0.45, 0.15, 0.35]
# pie chart
# def pie(
#     x: ArrayLike,
#     explode: ArrayLike | None = None,
#     labels: Sequence[str] | None = None,
#     colors: ColorType | Sequence[ColorType] | None = None,
#     autopct: str | Callable[[float], str] | None = None,
#     pctdistance: float = 0.6,
#     shadow: bool = False,
#     labeldistance: float | None = 1.1,
#     startangle: float = 0,
#     radius: float = 1,
#     counterclock: bool = True,
#     wedgeprops: dict[str, Any] | None = None,
#     textprops: dict[str, Any] | None = None,
#     center: tuple[float, float] = (0, 0),
#     frame: bool = False,
#     rotatelabels: bool = False,
#     *,
#     normalize: bool = True,
#     hatch: str | Sequence[str] | None = None,
#     data=None,
# ) -> tuple[list[Wedge], list[Text]] | tuple[list[Wedge], list[Text], list[Text]]:

# soldNums：表示饼图的数据，即每个扇区的数值。这可以是一个包含实数值的数组或序列。
# labels=kinds：设置饼图的标签，即每个扇区的标签。这里使用的标签是一个包含字符串的列表，每个字符串对应一个扇区的标签。
# autopct="%3.1f%%"：设置饼图的百分比标签格式。这里使用的格式字符串 "%3.1f%%" 指定了百分比的显示格式，保留一位小数。
# startangle=60：设置饼图的起始角度。这里将起始角度设置为 60 度，即第一个扇区的起始位置在水平方向上顺时针旋转 60 度。
# colors=colors：设置饼图的颜色。这里使用的颜色是一个包含颜色名称或颜色代码的列表，用于指定每个扇区的颜色。
plt.pie(soldNums, labels=kinds, autopct="%3.1f%%", startangle=60, colors=colors)
plt.title("不同类型箱子的销售数量占比")
plt.show()
```

#### 2.5 函数 polar()——用于绘制极线图

```python
import matplotlib.pyplot as plt
import numpy as np

barSlices = 12
# def linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None,axis=0):
# 0.0：起始角度，表示角度范围的起始值为 0 度。
# 2 * np.pi：结束角度，表示角度范围的结束值为 2π（360度），即一个完整的圆。
# barSlices：表示将圆分割为多少个扇区。
# endpoint=False：表示是否包含结束角度。设置为 False 表示不包含结束角度，即最后一个扇区的边界不与起始角度相连。
theta = np.linspace(0.0, 2 * np.pi, barSlices, endpoint=False)
r = 30 * np.random.rand(barSlices)

# def polar(*args, **kwargs) -> list[Line2D]:
# theta：表示极坐标图的角度值，即每个点在极坐标中的角度值。这可以是一个包含实数值的数组或序列。
# r：表示极坐标图的半径值，即每个点在极坐标中的半径值。这可以是一个包含实数值的数组或序列。
# color="chartreuse"：设置极坐标图的颜色。这里使用的颜色是 "chartreuse"，代表黄绿色。
# linewidth=2：设置极坐标图的线宽。这里将线宽设置为 2 像素。
# marker="*"：设置极坐标图的标记样式。这里使用的标记样式是星号（"*"），表示在每个点处绘制一个星号标记。
# mfc="b"：设置标记的填充色。这里将填充色设置为蓝色（"b"）。
# ms=10：设置标记的大小。这里将标记的大小设置为 10。
plt.polar(theta,
          r,
          color="chartreuse",
          linewidth=2,
          marker="*",
          mfc="b",
          ms=10)
plt.show()
```

#### 2.6 函数 scatter()——用于绘制气泡图

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

a = np.random.randn(100)
b = np.random.randn(100)
# colormap:RdYlBu
# a：表示散点图的 x 坐标值，即横坐标。这可以是一个包含实数值的数组或序列。
# b：表示散点图的 y 坐标值，即纵坐标。这可以是一个包含实数值的数组或序列。
# s=np.power(10 * a + 20 * b, 2)：设置散点的大小。这里根据公式 10 * a + 20 * b 计算每个点的值，并将其平方作为散点的大小。通过 np.power 函数对计算结果进行平方。
# c=np.random.rand(100)：设置散点的颜色。这里使用 np.random.rand(100) 生成一个包含 100 个随机值的数组，作为每个点的颜色值。
# cmap=mpl.cm.RdYlBu：设置颜色映射。这里使用的颜色映射是 mpl.cm.RdYlBu，表示使用 "RdYlBu" 色图来映射颜色值。
# marker="o"：设置散点的标记样式。这里使用的标记样式是圆形（"o"），表示在每个点处绘制一个圆形标记。
plt.scatter(a,
            b,
            s=np.power(10 * a + 20 * b, 2),
            c=np.random.rand(100),
            cmap=mpl.cm.RdYlBu,
            marker="o")
plt.show()
```

#### 2.7 函数 stem()——用于绘制棉棒图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.5, 2 * np.pi, 20)
y = np.random.randn(20)
# def stem(
#     *args: ArrayLike | str,
#     linefmt: str | None = None,
#     markerfmt: str | None = None,
#     basefmt: str | None = None,
#     bottom: float = 0,
#     label: str | None = None,
#     orientation: Literal["vertical", "horizontal"] = "vertical",
#     data=None,
# ) -> StemContainer:

# x：表示离散数据点的 x 坐标值，即横坐标。这可以是一个包含实数值的数组或序列。
# y：表示离散数据点的 y 坐标值，即纵坐标。这可以是一个包含实数值的数组或序列。
# linefmt="-."：设置线的样式。这里使用的线样式是点划线（"-."），表示在每个数据点处绘制一个垂直线段。
# markerfmt="o"：设置标记的样式。这里使用的标记样式是圆形（"o"），表示在每个数据点处绘制一个圆形标记。
# basefmt="-"：设置基线的样式。这里使用的基线样式是实线（"-"），表示绘制垂直线段的基线。
plt.stem(x, y, linefmt="-.", markerfmt="o", basefmt="-")
plt.show()
```

#### 2.9 函数 errorbar()——用于绘制误差棒图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)
# def errorbar(
#     x: float | ArrayLike,
#     y: float | ArrayLike,
#     yerr: float | ArrayLike | None = None,
#     xerr: float | ArrayLike | None = None,
#     fmt: str = "",
#     ecolor: ColorType | None = None,
#     elinewidth: float | None = None,
#     capsize: float | None = None,
#     barsabove: bool = False,
#     lolims: bool | ArrayLike = False,
#     uplims: bool | ArrayLike = False,
#     xlolims: bool | ArrayLike = False,
#     xuplims: bool | ArrayLike = False,
#     errorevery: int | tuple[int, int] = 1,
#     capthick: float | None = None,
#     *,
#     data=None,
#     **kwargs,
# ) -> ErrorbarContainer:

# x：表示散点图的 x 坐标值，即横坐标。这可以是一个包含实数值的数组或序列。
# y：表示散点图的 y 坐标值，即纵坐标。这可以是一个包含实数值的数组或序列。
# fmt="bo:"：设置数据点的样式。这里使用的样式是蓝色圆点加上横向冒号线（"bo:"），表示在每个数据点处绘制一个蓝色圆点，并用横向冒号线连接。
# yerr=0.2：设置纵向误差线的大小。这里将纵向误差线的大小设置为 0.2。
# xerr=0.02：设置横向误差线的大小。这里将横向误差线的大小设置为 0.02。
plt.errorbar(x, y, fmt="bo:", yerr=0.2, xerr=0.02)
plt.xlim(0, 0.7)
plt.show()
```
